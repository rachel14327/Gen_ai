```python
# PDF reader
from langchain_community.document_loaders import PyPDFLoader
loader = PyPDFLoader('../invoice.pdf')
docs = loader.load()
docs
```

```python
#Convert pdf document to chuncks
from langchain_text_splitters import RecursiveCharacterTextSplitter
text_splitter = RecursiveCharacterTextSplitter(chunk_size=1000, chunk_overlap=200)
chunks = text_splitter.split_documents(docs)
chunks
```

```python
# Convert the chunks to a vector store
from langchain_ollama import OllamaEmbeddings
from langchain_community.vectorstores import Chroma

db = Chroma.from_documents(
    documents=chunks,
    embedding=OllamaEmbeddings(model="nomic-embed-text")
)
```

```python
##Vector database
query = "What is the Total Amount Incl. VAT ?"
result = db.similarity_search(query)
result[0].page_content
```

```python
## FAISS vector store
from langchain_community.vectorstores import FAISS
Fais_db = FAISS.from_documents(chunks, embedding=OllamaEmbeddings(model="nomic-embed-text"))
```

```python
##FAISS database
query = "What is the Total Amount Incl. VAT ?"
result = Fais_db.similarity_search(query)
result[0].page_content
```

```python
from langchain_ollama import ChatOllama
# Load Ollama LLM model
llm = ChatOllama(model="llama3.1")
llm 
```

```python
## Designing chat prompt template
from langchain_core.prompts import ChatPromptTemplate
prompt = ChatPromptTemplate.from_template("""
Answer the question based on the context provided.
Think step by step before providing the detailed answer.
I will tip you 100$ if you answer correctly.
<context>
{context}
</context>
Question: {input}
""")
```

```python
from langchain_core.output_parsers import StrOutputParser
document_chain = prompt | llm | StrOutputParser()
```

```python
from langchain_core.runnables import RunnablePassthrough
retriever = Fais_db.as_retriever() # Retrieves the most relevant documents from the vector store
retrieval_chain = {
    "context": retriever,
    "input": RunnablePassthrough() # it will pass the input to LLM forward without any changes
} | document_chain
```

```python
response = retrieval_chain.invoke("What is the Total Amount Incl. VAT ?")
response['answer']
```
